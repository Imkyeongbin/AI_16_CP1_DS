```python
import os, json, glob
import pandas as pd

def find_files_to_df(directory):
    # 모든 JSON 파일을 저장할 빈 리스트 생성
    json_files = []
    for root, dirs, files in os.walk(directory):
        for file in glob.glob(os.path.join(root, f'*.json') ):
            with open(file, 'r') as f:
                json_data = f.read()
            # JSON 데이터 파싱 후 리스트에 추가
            parsed_json = json.loads(json_data)['FILE'][0]
            parsed_json['file_directory'] = root
            json_files.append(parsed_json)

    return pd.DataFrame(json_files)
```

```python
## 라벨 경로 설정
dir_path_labels = os.path.relpath('../../New_Sample/라벨링데이터')

df = find_files_to_df(dir_path_labels)
```

```python
df['IMG_WIDTH'] = df["RESOLUTION"].apply(lambda x: x.split('*')[0])
df['IMG_HEIGHT'] = df["RESOLUTION"].apply(lambda x: x.split('*')[1])
df['file_directory'] = df['file_directory'].apply(lambda x: x.replace('라벨링데이터\TL', '원천데이터\TS'))
df
```

```python
import shutil

def output_yolo(row, set_type):
    img_width = float(row['IMG_WIDTH'])
    img_height = float(row['IMG_HEIGHT'])
    filename = row['FILE_NAME'].rsplit('.', 1)[0] + '.txt'

    # 파일이 저장될 디렉토리 경로
    dir_path = os.path.relpath('../dataset')

    labels_path = os.path.join(dir_path, 'labels', set_type)
    images_path = os.path.join(dir_path, 'images', set_type)
    if not os.path.exists(labels_path):
        os.makedirs(labels_path)
    if not os.path.exists(images_path):
        os.makedirs(images_path)

    # 파일 전체 경로
    file_path = os.path.join(labels_path, filename)
    with open(file_path, 'w', encoding="utf-8") as f:
        for item in row['ITEMS']:
            class_index = 0  # 클래스 인덱스는 항상 0으로 설정
            if item['PACKAGE'] == '불법차량':
                class_index = 1
            x, y, w, h = map(float, item['BOX'].split(','))
            cx = x + w / 2
            cy = y + h / 2
            width = w / img_width
            height = h / img_height
            cx = cx / img_width
            cy = cy / img_height
            f.write(f"{class_index} {cx:.6f} {cy:.6f} {width:.6f} {height:.6f}\n")
        # 파일을 복사함
        shutil.copy(os.path.join(row['file_directory'], row['FILE_NAME']), os.path.join(images_path, row['FILE_NAME']))
```

```python
from sklearn.model_selection import train_test_split
df_train, df_val = train_test_split(df, test_size = 0.2, random_state=42)
```

```python
df_train.apply(lambda x: output_yolo(x, 'train'), axis=1)
df_val.apply(lambda x: output_yolo(x, 'val'), axis=1)
```
