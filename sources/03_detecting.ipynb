```python
import time, platform, sys
def finish_alarm():
    
    if platform.system() == 'Windows':
        import winsound
        for i in range(3):
            winsound.MessageBeep()
            time.sleep(2)
    elif platform.system() == 'Darwin':  # macOS
        import AppKit
        sound = AppKit.NSSound.alloc()
        sound.initWithContentsOfFile_byReference_('/System/Library/Sounds/Basso.aiff', True)
        sound.play()
    elif platform.system() == 'Linux':
        # 예시로 pygame 라이브러리를 사용하여 소리를 재생합니다.
        import pygame

        pygame.mixer.init()
        pygame.mixer.music.load('../alarm/alarm.wav')
        pygame.mixer.music.play()
    else:
        sys.stderr.write('Unsupported platform: %s' % platform.system())
```

```python
# yolov5/runs/train의 exp 마지막 디렉토리에 있는 weights/best.pt를 구한다.
import os
def find_lastexp():
    dir_prefix = 'exp'
    max_dir_num = -1

    for dir_name in os.listdir('../yolov5/runs/train'):
        if dir_name.startswith(dir_prefix):
            try:
                dir_num = int(dir_name[len(dir_prefix):])
                if dir_num > max_dir_num:
                    max_dir_num = dir_num
            except:
                # exp'' 인 경우
                max_dir_num = 0

    # max_dir_num 값에 따라 max_dir_name을 대입한다.
    max_dir_name = dir_prefix + str(max_dir_num if max_dir_num != 0 else '')
    return max_dir_name
```
